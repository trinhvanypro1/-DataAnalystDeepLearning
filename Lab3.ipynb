1. Missing header

```python
import pandas as pd
data=pd.read_csv('https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/patient_heart_rate.csv')
data.head(10)
```

a. Add columns name for dataframe with information such as: (ID,Name,Age,Weight,m0006,m0612,m1218,f006,f0612,f1218)

```python

data=pd.read_csv('https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/patient_heart_rate.csv',names=['ID','Name','Age','Weight','m0006','m0612','m1218','f006','f0612','f1218'])
data.head(6)
```

2. Multi-variables in one columns

```python
data[['First_Name','Last_Name']]=data['Name'].str.split(' ',expand=True)
```

```python
data.head(10)
```

Delete columns "Name":

```python
del data['Name']
```

```python
data.head(10)
```

3. Inconsitence values.
Convert "lbs" to "kg" with columns name "Weight".

```python
weight=data['Weight']
for i in range(len(weight)):
  x=str(weight[i])
  if "lbs" in x[-3:]:
    x=x[:-3:]
    y=int(float(x)/2.2)
    y=str(y)+'kgs'
    weight[i]=y
```

```python
data.head(10)
```

4– Drop An empty row[8,9] in the data

```python
data.dropna(how='all',inplace=True)
```

```python
data
```

5.Duplicate records in the data

In data, there is repeat in 6 and 9 row.

```python
data=data.drop_duplicates(subset=['Age','Weight','First_Name','Last_Name'])
data
```

6. Inconsitence characters (error ASCII-based format)

Error ASCII-based format in the Fisrt_Name and Last_Name columns in the data.

```python
data.First_Name.replace({r'[^\x00-\x7F]+':''},regex=True,inplace=True)
data.Last_Name.replace({r'[^\x00-\x7F]+':''},regex=True,inplace=True)
data
```

7. Missing values

a.Check for empty data

```python
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format= 'svg'
sns.heatmap(data.isna(),
            yticklabels=False,
            cmap='cool')
```

b. Get the index of rows with missing information Age and Weight.

```python
index_age_weight_null=data[data['Age'].isna() & data['Weight'].isna()].index
index_age_weight_null
```

c. Drop index Age & Weight are null

```python
data.drop(index_age_weight_null,inplace=True)
```

d. Check data once again

```python
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format='svg'
sns.heatmap(data.isna(),
            yticklabels=True,
            cbar=True,
            cmap='cool')
```

```python
data['Age'].fillna(data['Age'].mean(),inplace=False)
```

8. Missing headers

a. Melt the Sex + time range columns in single column

```python
data=pd.melt(data,id_vars=['ID', 'Age', 'Weight', 'First_Name', 'Last_Name'],value_name='PulseRate',var_name='sex_and_time').sort_values(['ID','Age','Weight','First_Name','Last_Name'])
data.head(10)
```

b. Extract to " Sex,Hour_Lower,Hour_Upper" from sex_and_time columns:


```python
tmp_data=data['sex_and_time'].str.extract('(\D)(\d+)(\d{2})',expand=True)
```

c. Create "Sex, Hour_Lower, Hour_Upper" columns

```python
tmp_data.columns=['Sex','Hour_Lower','Hour_Upper']
```

d.Create "Time" columns based on "Hour_Lower, Hour_Upper" columns

```python
tmp_data['Time']=tmp_data['Hour_Lower']+'-'+ tmp_data['Hour_Upper']
```

```python
#Merge "data and tmp_data":
data=pd.concat([data,tmp_data],axis=1)
```
