Thống kê dữ liệu thiếu cho cột dân tộc và hiệu chỉnh dữ liệu thiếu như sau: 

Mặc định thiếu thì điền giá trị 0

```python
data_XT['DT'].fillna(0, inplace= True)
```

```python
data_XT['DT'].head(10)
```

Thống kê dữ liệu thiếu cho biến T1 và hiệu chỉnh dữ liệu, lưu ý việc thay thế dữ liệu thiếu sử dụng phương pháp Mean.

```python
data_XT['T1'].fillna(data_XT['T1'].mean(),inplace=True)
```

```python
data_XT.head(10)
```

Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và 12. 

• Công thức tính: TBM = (T2 + L + H + S + V2 + X + D + N) / 10

```python
for i in ['1','2','3']:
  data_XT['TBM'+i] = (data_XT['T'+i]*2 + data_XT['L'+i] + data_XT['H'+i] + data_XT['S'+i] + data_XT['V'+i]*2 + data_XT['X'+i] + data_XT['D'+i] + data_XT['N'+i])/10
```

```python
data_XT.head(4)
```

Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1, TBM2 và TBM3 cho từng năm lớp 10, 11, 12 như sau: 

• Nhỏ hơn 5.0 xếp loại: yếu (kí hiệu là Y) 

• Từ 5.0 đến dưới 6.5: trung bình (kí hiệu là TB) 

• Từ 6.5 đến dưới 8.0: khá (kí hiệu là K) 

• Từ 8.0 đến dưới 9.0: giỏi (kí hiệu là G) 

• Từ 9.0 trở lên: xuất sắc (kí hiệu là XS)

```python
for i in ['1','2','3']:
  data_XT.loc[(data_XT['TBM'+i]) < 5, 'XL'+i ]='Y'
  data_XT.loc[(data_XT['TBM'+i]>=5)&(data_XT['TBM'+i]<6.5),'XL'+i]='TB'
  data_XT.loc[(data_XT['TBM'+i]>=6.5)&(data_XT['TBM'+i]<8),'XL'+i]='K'
  data_XT.loc[(data_XT['TBM'+i]>=8)&(data_XT['TBM'+i]<9),'XL'+i]='G'
  data_XT.loc[(data_XT['TBM'+i]>=9),'XL'+i]='XS'
```

```python
data_XT.head(5)
```

Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp 10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ. Sử dụng phương pháp Min-Max Normalization

```python
for i in ['1','2','3']:
  data_XT['US_TBM'+i] = (data_XT['TBM'+i]/10)*4
```
