Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và 12. • Công thức tính: TBM = (T2 + L + H + S + V2 + X + D + N) / 10

```python
for i in['1','2','3']:
  print(['TBM'+i])
```

```python
for i in ['1','2','3']:
  data['TBM'+i]=(data['T'+i]*2+data['L'+i]+data['H'+i]+data['S'+i]+data['V'+i]*2+data['X'+i]+data['D'+i]+data['N'+i])/10
  
```

Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1, TBM2 và TBM3 cho từng năm lớp 10, 11, 12 như sau: • Nhỏ hơn 5.0 xếp loại: yếu (kí hiệu là Y) • Từ 5.0 đến dưới 6.5: trung bình (kí hiệu là TB) • Từ 6.5 đến dưới 8.0: khá (kí hiệu là K) • Từ 8.0 đến dưới 9.0: giỏi (kí hiệu là G) • Từ 9.0 trở lên: xuất sắc (kí hiệu là XS)

```python
for i in ['1','2','3']:
  data.loc[(data['TBM'+i]<5),'XL'+i]='Y'
  data.loc[(data['TBM'+i]>=5)&(data['TBM'+i]<6.5),'XL'+i]='TB'
  data.loc[(data['TBM'+i]>=6.5)&(data['TBM'+i]<8),'XL'+i]='K'
  data.loc[(data['TBM'+i]>=8)&(data['TBM'+i]<9),'XL'+i]='G'
  data.loc[(data['TBM'+i]>=9),'XL'+i]='XS'
```

Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp 10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ. Sử dụng phương pháp Min-Max Normalization

```python
for i in ['1','2','3']:
  data['US_TMN'+i]=(data['TBM'+i]/10)*4
print(data)
```

Tạo biến kết quả xét tuyển (kí hiệu là KQXT) nhằm xác định:

- Sinh viên đậu (giá trị “1”) và rớt (giá trị “0”)

Các khối dựa trên điểm DH1, DH2 và DH3 như sau:

• Với khối A, A1 nếu [(DH12 + DH2 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu, ngược lại là rớt 

• Với khối B nếu [(DH1 + DH22 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu, ngược lại là rớt 

• Với khối khác nếu [(DH1+ DH2 + DH3)/3] lớn hơn hoặc bằng 5.0 thì đậu, ngược lại là rớt

```python
for i in data['KT'].unique():
  data.loc[(i=='A'or i=='A1')&(((data['DH1']*2+data['DH2']+data['DH3'])/4) <5),'KQXT']=0
  data.loc[(i=='A'or i=='A1')&(((data['DH1']*2+data['DH2']+data['DH3'])/4) >=5),'KQXT']=1
  
  data.loc[(i=='B')&(((data['DH1']+data['DH2']*2+data['DH3'])/4) <5),'KQXT']=0
  data.loc[(i=='B')&(((data['DH1']+data['DH2']*2+data['DH3'])/4) >=5),'KQXT']=1

  data.loc[(i!='A1' and i!='A' and i!='B') & ((data['DH1'] +data['DH2']+ data['DH3'])/3 >=5), 'KQXT']=1
  data.loc[(i!='A1' and i!='A' and i!='B') & ((data['DH1'] +data['DH2']+ data['DH3'])/3 <5), 'KQXT']=0
```
